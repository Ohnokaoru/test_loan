### 每個月還款金額利息

```python
datas=[]

amount=12000000
years=30
rate=2.06 #年利率

month_rate=rate/100/12 # 月利率

period=years*12 #總期數
month_pay=int(amount/period) #每月應付本金

totalInterest=0 
for i in range(period):
    interest=round(amount*month_rate) #月利息
    amount-=month_pay
    totalInterest+=interest #總支出利息

    if i==period-1:
        datas.append([i+1,month_pay+amount,interest,month_pay+interest+amount,0])
        print(f"{i+1}期，本金:{month_pay+amount}，利息:{interest}，本利和:{month_pay+interest+amount}，餘額:0")
    else:
         print(f"{i+1}期，本金:{month_pay}，利息:{interest}，本利和:{month_pay+interest}，餘額:{amount}")
         datas.append([i+1,month_pay,interest,month_pay+interest,amount])
  

#總支出利息
print(totalInterest)

```
