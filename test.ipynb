### 每個月還款金額利息

```python
datas=[]
total_amount=100000
amount=total_amount #每月還本金會遞減
years=5
rate=1 #年利率

month_rate=rate/100/12 # 月利率

period=years*12 #總期數
month_pay=int(amount/period) #每月應付本金

totalInterest=0 
for i in range(period):
    interest=round(amount*month_rate) #月利息
    amount-=month_pay
    totalInterest+=interest #總支出利息

    if i==period-1:
        datas.append([i+1,month_pay+amount,interest,month_pay+interest+amount,0])
        print(f"{i+1}期，本金:{month_pay+amount}，利息:{interest}，本利和:{month_pay+interest+amount}，餘額:0")
    else:
         print(f"{i+1}期，本金:{month_pay}，利息:{interest}，本利和:{month_pay+interest}，餘額:{amount}")
         datas.append([i+1,month_pay,interest,month_pay+interest,amount])
  

    #總支出利息
    print(totalInterest)
    #總還款金額
    print(totalInterest+total_amount)
```

# 計算方式：
試算公式：
每月應付本息金額之平均攤還率 ＝{[(1＋月利率)^月數]×月利率}÷{[(1＋月利率)^月數]－1}
(公式中：月利率 ＝ 年利率／12 ； 月數=貸款年期 ｘ 12)
每月應攤還本金與利息試算：
平均每月應攤付本息金額＝貸款本金×每月應付本息金額之平均攤還率 ＝每月應還本金金額＋每月應付利息金額
每月應付利息金額＝本金餘額×月利率
每月應還本金金額＝平均每月應攤付本息金額－每月應付利息金額

```python
import math

total_amount=1000000
amount=total_amount

years=5
rate=2


month_rate=rate/100/12 
period=years*12

month_pay=amount*month_rate*(pow((1+month_rate),period))/(pow((1+month_rate),period)-1)
month_pay=math.ceil(month_pay)

datas=[]
totalInterest=0

for i in range(period):
    pay_month_interest=amount*month_rate
    pay_month_interest=round(pay_month_interest)
    amount-=(month_pay-pay_month_interest)
    totalInterest+=pay_month_interest
    

    if i!=period-1:
        print(f"期別:{i+1},本金:{month_pay-pay_month_interest},利息:{pay_month_interest},本利和:{month_pay},餘額:{amount}")
        datas.append([i+1,month_pay-pay_month_interest,pay_month_interest,month_pay,amount])
    else:
        print(f"期別:{i+1},本金:{amount+month_pay-pay_month_interest},利息:{pay_month_interest},本利和:{amount+month_pay},餘額:{0}")
        datas.append([i+1,amount+month_pay-pay_month_interest,pay_month_interest,amount+month_pay,0])

#總支出利息
print(totalInterest)
#總還款金額
print(totalInterest+total_amount)
datas
```
